### Model Framework : Spacy
Model trained on the given corpus to retrieve labels from clinical based texts. Although the corpus contains 5 labels we have emphasised on **Drug** and **ADR** which is the crucial information on drug safety. The NER model can be further used different NLP based applications. This is the first ever open sourced model on ADR and Drug. 


Import Libraries

```python
import spacy
from spacy import displacy
```

```python
PATH = './spacy_adr'
```

Load custom ner model

```python
nlp = spacy.load(PATH)
```

Set colors for each entities

```python
colors = {"Drug": "linear-gradient(90deg, #aa9cfc, #fc9ce7)", "ADR":"linear-gradient(60deg, #ffdf80, #ffffed)","Disease":"linear-gradient(60deg, #add8e6, #90ee90)"}
options = {"ents": ["ADR","Disease","Drug"], "colors": colors}
```

### Example 1

```python
text = '''A 46-year-old female patient with a history of a sleeve gastrectomy started complaining about episodes of lipothymias preceded by sweating, nausea, and dizziness. During one of these episodes, a capillary blood glucose was obtained with a value of 24 mg/dl. She had multiple emergency admissions with low-capillary glycemia. An exhaustive investigation for possible causes of hypoglycemia was made for 18 months. The 72h fasting test was negative for hypoglycemia. A Raynaud’s phenomenon was identified during one appointment.'''
doc = nlp(text)
displacy.render(doc,style = "ent",options=options)
```

### Example 2

```python
text = '''Isotretinoin is considered the gold standard treatment for severe nodulocystic acne, though it has been the subject of controversy in the media for concerns related to adverse psychiatric effects. In 2005, the FDA issued a black box warning for suicide, depression, aggression, and psychosis.However, repeated high quality studies have found no association between isotretinoin and psychiatric disorders.'''
doc = nlp(text)
displacy.render(doc, style="ent",options=options)
```
